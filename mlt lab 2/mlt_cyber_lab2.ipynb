```python
import pandas as pd

# Load the dataset into a Pandas DataFrame
file_path = "real_estate.csv"
df = pd.read_csv(file_path)

# Create a unique identifier for each row (replace the address with a unique number)
df['UniqueID'] = range(1, len(df) + 1)
df.drop('Address', axis=1, inplace=True)  # Drop the original address column

# Define the features and target variable
features = ['Rooms', 'Price', 'Bedroom2', 'Bathroom', 'Landsize', 'YearBuilt', 'Regionname', 'Suburb']
target_variable = 'Type'


# Map the target variable 'Type of House' to 't'
df['Type'] = df['Type'].map({'h': 'house, cottage, villa, semi, terrace',
                                               'u': 'unit, duplex', 't': 'townhouse'})

# Display the modified DataFrame
print(df[features + [target_variable]])
```

```python
# Define the bins and labels for categorizing "Year Built"
bins = [-float('inf'), 1800, 1850, 1900, 1920, 1950, 1980, 2000, 2023, float('inf')]
labels = [8, 7, 6, 5, 4, 3, 2, 1, 0]

# Apply the categorization to the "Year Built" column
df['YearBuilt'] = pd.cut(df['YearBuilt'], bins=bins, labels=labels, include_lowest=True)
```

```python
# Import necessary libraries for normalization
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler
scaler = MinMaxScaler()

# Normalize "Price" and "Land Size"
df[['Price', 'Landsize']] = scaler.fit_transform(df[['Price', 'Landsize']])
```

## Covariance matrix
Create a numpy array with 3 features land size, price, number of bedrooms. the size of the array will be 3xN. N is the number of samples.

An example of how to find the covariance matrix is given below:

```python
import numpy as np

# Assuming df is your DataFrame with the required features
features = ['Landsize', 'Price', 'Bedroom2']
subset_df = df[features].dropna()  # Drop rows with missing values, if any

# Create a numpy array with the selected features
data_array = subset_df.to_numpy().T  # Transpose to get a 3xN array

# Calculate the covariance matrix
covariance_matrix = np.cov(data_array)

# Display the covariance matrix
print("Covariance Matrix:")
print(covariance_matrix)
```

## Basic linear algebra
- Find the eigen values of the covariance matrix of the entire dataset (use eigen value decomposition)
- Find the singular values of the covariance matrix of the entire dataset

```python
import numpy as np
import pandas as pd

# Load the real estate dataset into a Pandas DataFrame
file_path = "real_estate.csv"
df = pd.read_csv(file_path)

# Select numeric columns for eigenvalue and singular value calculation
numeric_columns = df.select_dtypes(include=[np.number])

# Drop any rows with missing values
numeric_columns = numeric_columns.dropna()

# Calculate the covariance matrix
covariance_matrix = np.cov(numeric_columns, rowvar=False)

# Eigenvalues and eigenvectors using eigenvalue decomposition
eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)

# Singular values using singular value decomposition
singular_values = np.linalg.svd(covariance_matrix, compute_uv=False)

# Display the results
print("Eigenvalues:")
print(eigenvalues)

print("\nSingular Values:")
print(singular_values)
```
