```python
def plot_iris(iris, col1, col2):
    plt.scatter(iris.data[:, col1], iris.data[:, col2], c=iris.target)
    plt.xlabel(iris.feature_names[col1])
    plt.ylabel(iris.feature_names[col2])
    plt.title(f'{iris.feature_names[col1]} vs {iris.feature_names[col2]}')
    plt.show()

plot_iris(iris, 0, 1)
plot_iris(iris, 0, 2)
plot_iris(iris, 0, 3)
plot_iris(iris, 1, 2)
plot_iris(iris, 1, 3)
plot_iris(iris, 2, 3)
```

```python

# Preprocessing Input data
data = load_iris()

X = data.data[:, 0]
Y = data.data[:, 2]

m = 0
c = 0

L = 0.001  # The learning Rate

# Set a threshold for the change in parameters to determine when to stop
threshold = 0.0000001

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
while True: 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m_new = m - L * D_m  # Update m
    c_new = c - L * D_c  # Update c
    
    # Check the difference between new and old parameter values
    if abs(m_new - m) < threshold and abs(c_new - c) < threshold:
        break  # Stop the iteration if the difference is negligible
    
    m = m_new
    c = c_new
    
print(m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot(X, Y_pred, color='yellow')  # regression line
plt.show()
```
